```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Total_Amount'].fillna(df['Total_Amount'].mean(), inplace=True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Total_Amount']= df['Total_Amount'].astype(int)
```

```python
df.to_csv('Sales_data.csv', index=False)
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total_Amount'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Total_Amount'], marker='o')
plt.title('Sales Performance Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby('Category')['Total_Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Total_Amount', data=category_sales)
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', hue='Region', data=df, palette='Set2')
plt.title('Number of Transactions by Product Category and Region', fontsize=16, fontweight='bold')
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.legend(title='Region')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

category_region_counts = df.groupby(['Category', 'Region']).size().reset_index(name='Transactions')


plt.figure(figsize=(10, 6))

category_data = category_region_counts.groupby('Category')['Transactions'].sum()
plt.pie(category_data, labels=category_data.index, autopct='%1.1f%%', colors=plt.get_cmap('Set2').colors)

plt.title('Number of Transactions by Product Category', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
